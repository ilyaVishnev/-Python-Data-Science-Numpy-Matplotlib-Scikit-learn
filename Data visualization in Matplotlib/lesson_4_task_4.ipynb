```python
import pandas as ps
import matplotlib.style as style
from matplotlib import pyplot as plt


df=ps.read_csv('D:\creditcard.csv', index_col=0)

df.head(10)
```

```python
style.use('fivethirtyeight')
```

```python
classes=df['Class'].value_counts()
classes.plot(kind="bar")
```

```python
classes.plot(kind="bar", logy=True)
```

```python
y1=df.loc[df["Class"]==0, "V1"]
y2=df.loc[df["Class"]==1, "V1"]

plt.hist([y1,y2], alpha=0.5, density=True, bins=20, color=['red','grey'])

plt.xlabel('Class')
plt.legend(labels=['Class 1', 'Class 0'])
```
